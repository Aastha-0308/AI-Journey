```python
df.sort_values("members", ascending=False)[["title","title_english", "members"]].head(10)
top10 = df.sort_values("members", ascending=False).head(10)

plt.figure(figsize=(20,5))
plt.bar(top10["title_english"], top10["members"])
plt.xticks(rotation=45)
plt.title("Top 10 Anime by Members")
plt.show()
```

```python
all_genres = df["genre"].dropna().str.split(", ")
# all_genres.head(10)
genre_counts = {}

for genres in all_genres:
    for genre in genres:
        genre_counts[genre] = genre_counts.get(genre, 0) + 1

sorted_genres = sorted(
    genre_counts.items(),
    key=lambda x:x[1],
    reverse=True
)

sorted_genres[:10]
```
