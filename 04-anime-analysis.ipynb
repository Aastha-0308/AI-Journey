```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("AnimeList.csv")
print("loaded")
```

```python
df_clean = df[df["score"] > 0]
```

```python
df_clean["members"].corr(df_clean["score"])
```

```python
df_clean["favorites"].corr(df_clean["score"])
```

```python
df_clean["episodes"].corr(df_clean["score"])
```

```python
df_clean.plot.scatter(x="episodes", y="score")
```
